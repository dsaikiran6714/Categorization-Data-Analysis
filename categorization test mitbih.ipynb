```python
X_test = mitbih_test.drop(columns=['feature188'])
y_test = mitbih_test['feature188']
```

```python
import joblib
loaded_data = joblib.load('mitbih_train')
```

```python
logistic_model = loaded_data['logistic_model']
tree_model = loaded_data['tree_model']
forest_model = loaded_data['forest_model']
```

```python
logistic_pred = logistic_model.predict(X_test)
tree_pred = tree_model.predict(X_test)
forest_pred = forest_model.predict(X_test)
```

```python
# Define model names for plotting
model_names = ['Logistic Regression', 'Decision Tree', 'Random Forest']
predictions = [logistic_pred, tree_pred, forest_pred]
```

```python
# Plotting results
plt.figure(figsize=(12, 8))
for i, pred in enumerate(predictions):
    plt.subplot(2, 3, i+1)
    plt.title(model_names[i])
    plt.scatter(range(len(pred)), y_test, label='True', color='blue', marker='.')
    plt.scatter(range(len(pred)), pred, label='Predicted', color='red', marker='x')
    plt.legend()
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
# Print evaluation results
for i, pred in enumerate(predictions):
    model_name = model_names[i]
    accuracy = accuracy_score(y_test, pred)
    cm = confusion_matrix(y_test, pred)
    report = classification_report(y_test, pred)
    
    print(f"Model: {model_name}")
    print(f"Accuracy: {accuracy:.4f}")
    print("Confusion Matrix:")
    print(cm)
    print("Classification Report:")
    print(report)
    print("-----------------------------------")
plt.tight_layout()
plt.show()
```
